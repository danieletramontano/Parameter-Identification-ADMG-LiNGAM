# How to use the identification algorithms on a fixed graph

```python
from identification import check_criterion, wholematrix_criterion, rand_graphs_sample
import warnings
warnings. filterwarnings('ignore')
```

## Define the graph and the parameters of interest 

```python
n = 30 # number of nodes
m = 10 # number of edges
g_dir, g_bid = rand_graphs_sample(n = n, m = m) # simulated random graph

v = 5 # target node of interest
q = {1, 2, 3, 4} # source nodes of interest
```

```python
id_q = check_criterion(g_dir, g_bid, v, q)
```

## Test identifiability of the parameters defined above

```python
if id_q:
    print("The vector of direct causal effects from ", q, "to ", v, "is IDENTIFIABLE")
else:
    print("The vector of direct causal effects from ", q, "to ", v, "is NOT IDENTIFIABLE")
```

## Test identifiability of the adjacency matrix

```python
id_matrix = wholematrix_criterion(g_dir, g_bid)
```

```python
if id_matrix:
    print("The adjecency matrix is IDENTIFIABLE")
else:
    print("The adjecency matrix is NOT IDENTIFIABLE")
```
